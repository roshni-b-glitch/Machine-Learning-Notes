```python
#LABEL ENCODING
from sklearn.preprocessing import LabelEncoder
Label_Encoder=LabelEncoder()
```

```python
sal_data1['Gender_Encode'] = Label_Encoder.fit_transform(sal_data1['Gender'])
```

```python
sal_data1['Degree_Encode'] = Label_Encoder.fit_transform(sal_data1['Degree'])
```

```python
sal_data1['Job_Titile_Encode'] = Label_Encoder.fit_transform(sal_data1['Job_Titile'])
```

```python
#data after label encoding
sal_data1.head()
```

```python
#feature scaling
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
sal_data1['Age_Scaled'] = std_scaler.fit_transform(sal_data1[['Age']])
sal_data1['Experience_Years_Scaled'] = std_scaler.fit_transform(sal_data1[['Experience_Years']])
```

```python
#data afer scaaling
sal_data1.head()
```

```python
#dependent and independent features
X=sal_data1[['Age_Scaled','Gender_Encode','Degree_Encode','Experience_Years_Scaled','Job_Titile_Encode']]
Y=sal_data1[['Salary']]
```

```python
#spliting the data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#model development
from sklearn.linear_model import LinearRegression
```

```python
Linear_regression_model=LinearRegression()
```

```python
#model training
Linear_regression_model.fit(X_train,Y_train)
```

```python
#model prediction
y_pred_lr=Linear_regression_model.predict(X_test)
y_pred_lr
```

```python
df=pd.DataFrame({'y_Actual':Y_test['Salary'],'y_Predicted':y_pred_lr.flatten()})
df
#
```

```python
df['Error']=df['y_Actual']-df['y_Predicted']
df
```

```python
#calculating absolute error
df['Absolute_Error']=np.absolute(df['Error'])
df
```

```python
#calculating mean absolute error
MAE=np.mean(df['Absolute_Error'])
MAE
```

```python
#model evaluation
from sklearn.metrics import accuracy_score,r2_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
#model accuracy
r2_score(Y_test,y_pred_lr)
```

```python
print('Mean Absolute Error:',mean_absolute_error(Y_test,y_pred_lr))
```

```python
#mean square error
mse=round(mean_squared_error(Y_test,y_pred_lr),2)
mse
```
